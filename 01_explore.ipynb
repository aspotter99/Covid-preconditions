Data from https://www.kaggle.com/tanmoyx/covid19-patient-precondition-dataset?select=covid.csv
this is a dataset with individual (de-identified) patient information of covid cases published by the Mexican government.

#### column descriptions:

- UPDATE_DATE The database is fed daily, this variable allows identifying the date of the last update.
- RECORD_ID Case identifier number
- ORIGIN Sentinel surveillance is carried out through the system of health units that monitor respirat...
- SECTOR Identifies the type of institution of the National Health System that provided care.
- ENTIDAD_UM Identifies the entity where the medical unit that provided the care is located.
- SEX Identifies the sex of the patient; 1= Female; 2 = Male
- ENTIDAD_NAC Identifies the patient's birth entity.
- ENTIDAD_RES Identifies the entity of residence of the patient.
- MUNICIPIO_RES Identifies the municipality of residence of the patient.
- PATIENT_TYPE Identifies the type of care received by the patient in the unit. 1 = Inpatient; 2 = Outpatient
- ENTRY_DATE Identifies the date of the patient's admission to the care unit.
- DATE_SYMPTOMS Identifies the date on which the patient's symptoms began.
- DATE_DEF Identifies the date the patient died. - 9999/99/99 = patient not deceased
- INTUBED Identifies if the patient required intubation. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- PNEUMONIA Identifies if the patient was diagnosed with pneumonia. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- AGE Identifies the age of the patient.
- NATIONALITY Identify if the patient is Mexican or foreign.
- PREGNANCY Identifies if the patient is pregnant. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- HABLA_LENGUA_INDIG Identifies if the patient speaks indigenous language.
- DIABETES Identifies if the patient has a diagnosis of diabetes. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- COPD Identifies if the patient has a diagnosis of COPD. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- ASMA Identifies if the patient has a diagnosis of asthma. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- INMUSUPR Identifies if the patient has immunosuppression. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- HYPERTENSION Identifies if the patient has a diagnosis of hypertension. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- OTRAS_COM Identifies if the patient has a diagnosis of other diseases. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- CARDIOVASCULAR Identifies if the patient has a diagnosis of cardiovascular disease. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- OBESITY Identifies if the patient is diagnosed with obesity. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- RENAL_CHRONIC Identifies if the patient has a diagnosis of chronic kidney failure. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- TOBACCO Identify if the patient has a smoking habit. 1 = Yes; 2 = No; 97,98,99 = missing or NA
- OTHER_CASE Identify if the patient had contact with any other case diagnosed with SARS CoV-2 1 = Yes; 2 = No; 97,98,99 = missing or NA
- RESULT Identifies the result of the analysis of the sample reported by the laboratory of the Nationa... 1 = Yes; 2 = No; 3 = Results Pending
- MIGRANT Identifies if the patient is a migrant. 
- PAIS_NACIONALIDAD Identifies the nationality of the patient.
- PAIS_ORIGEN Identifies the country from which the patient left for Mexico.
- ICU Identifies if the patient required to enter an Intensive Care Unit. 1 = Yes; 2 = No; 97,98,99 = missing or NA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
#reading the full data file into a dataframe.
data = pd.read_csv('covid.csv')
data
```

```python
# filtering - dataframe of only patient records where diabetes is present 

db = data[data.diabetes.eq(1)]

db
```

```python
def data_summary(df):
    eda_df= {}
    eda_df['null_sum'] = df.isnull().sum()
    eda_df['dtypes'] = df.dtypes
    eda_df['count'] = df.count()
    eda_df['mean'] = df.mean()
    eda_df['median'] = df.median()
    eda_df['max'] = df.max()
    eda_df['min'] = df.min()
    
    return pd.DataFrame(eda_df)

data_summary(db)
```
